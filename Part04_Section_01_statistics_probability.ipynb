## Part 04. 통계와 확률
확률 및 통계의 핵심개념과 다양한 확률분포를 다룸.
- 기초 확률 및 통계 개념 : 모집단과 표본, 표본추출 방법, 확률의 기본 성질
- 확률변수와 확률분포 : 이산형 및 연속형 확률변수, 확률질량함수(PMF), 확률밀도함수(PDF), 누적분포함수(CDF)
- 이산 확률분포 : 베르누이, 이항, 포아송 분포 및 Python을 활용한 확률 계산
- 연속 확률분포 : 정규, 카이제곱, t, F 분포의 개념과 활용
- Python을 활용한 확률분포 시뮬레이션 : Scipy를 이용한 확률 계산 및 샘플링

```python
# 기대값 E(X) 계산하기
import numpy as np

values = np.array([1, 2, 3, 4])
probabilities = np.array([0.1, 0.3, 0.2, 0.4])

print(np.sum(values * probabilities).round(3))
```

```python
# 분산 V(X) 계산하기.
# Var(X) = np.sum(xi - E(X))**2 * p(xi)
values = np.array([1, 2, 3, 4])
probabilities = np.array([0.1, 0.3, 0.2, 0.4])

print(np.sum((values - 2.9)**2 * probabilities).round(3))
```

누적분포함수는 `np.cumsum(probabilities)`으로 구할 수 있음.   
기대값은 `np.sum(value * prob)`   
분산은 `np.sum((xi - E(X))**2 * p(xi))` or `E(x**2) - (E(x)**2)`

### Section 01 연습문제 : 기초통계 & 확률의 이해

```python
#1 다음과 같은 확률질량함수를 따르는 이산 확률변수 X에 대해 x = 2일 때 누적분포함수 F(x)를 구하시오.

x = np.array([0, 1, 2, 3])
probs = np.array([0.1, 0.3, 0.4, 0.2])

cdf = np.cumsum(probs)
F_2 = cdf[2]

print(f"F(2) = {F_2}")
```

```python
#2 이산 확률변수 X에 대해 P(X = 2)와 P(X > 2)를 누적분포함수 F(x)를 이용해 구하시오.

x = np.array([0, 1, 2, 3])
probs = np.array([0.1, 0.3, 0.4, 0.2])

cdf = np.cumsum(probs)
F_2 = cdf[2] - cdf[1]
F_upper_2 = 1.0 - cdf[2]

print(f"F(2) = {F_2}")
print(f"F(x > 2) = {F_upper_2}")
```

```python
#3 확률밀도함수 f(x) = 1 / 3 (단, 0 <= x <= 3)인 확률변수 x에 대해 누적분포함수 F(x)를 정의하고 F(2)를 구하시오.

def F(x):
    if x >= 0 and x <= 3:
        return x/3
    elif x < 0:
        return 0
    else:
        return 1

print(F(2))
```

```python
#4 확률변수 X에 대한 확률질량함수가 다음과 같을 때, X의 기대값과 누적분포 함수 F(2)를 구하시오.

x = np.array([1, 2, 3])
probs = np.array([0.2, 0.5, 0.3])

cdf = np.cumsum(probs)
Ex = sum(x * probs)

F_2 = cdf[1]

print(f"기대값 = {Ex}")
print(f"F(2) = {F_2}")
```

```python
#5 확률변수 X의 확률분포가 다음과 같을 때, X의 기댓값과 분산을 구하시오.

x = np.array([0, 1, 2, 3])
probs = np.array([0.1, 0.3, 0.4, 0.2])

Ex = np.sum(x * probs)
# 분산은 Var(X) = (xi - E(X))**2 * p(xi) 로도 계산할 수 있지만, E(x**2) - (E(x)**2)으로 구할 수도 있음.
Ex_2 = np.sum(x**2 * probs)
Var = Ex_2 - Ex**2

print(f"기대값 = {Ex}")
print(f"분산 = {Var}")
```
