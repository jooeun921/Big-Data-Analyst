###### ğŸš©1ë²ˆ

```python
#1 ê³„ì ˆ(season)ì´ 1ì¼ ë•Œ, ì‹œê°„ëŒ€(hour)ë³„ ì´ ëŒ€ì—¬ëŸ‰(count í•©ê³„)ì„ ê³„ì‚°í•˜ê³ , ê·¸ ì¤‘ ëŒ€ì—¬ëŸ‰ì´ ê°€ì¥ ë§ì€ ì‹œê°„ëŒ€ë¥¼ êµ¬í•˜ì‹œì˜¤.

df_bike['datetime'] = pd.to_datetime(df_bike['datetime'])
df_bike['hour'] = df_bike['datetime'].dt.hour

df_hour_count = df_bike.loc[df_bike['season'] == 1, ['hour', 'count']]
df_hour_count_group = df_hour_count.groupby('hour').sum().reset_index()

max_hour = df_hour_count_group['count'].idxmax()
max_count = df_hour_count_group['count'].max()
print("ì‹œê°„ëŒ€:", max_hour, "ì‹œ ëŒ€ì—¬ëŸ‰: ", max_count , "ëŒ€", sep='')

# print(df_hour_count_group.loc[df_hour_count_group['count'].idxmax()])
```

```python
#7 ê³µìœ  ìì „ê±° ë°ì´í„°ë¥¼ í™œìš©í•˜ì—¬ ë„“ì€ í˜•ì‹(wide format)ì—ì„œ ê¸´ í˜•ì‹ìœ¼ë¡œ(long format)ìœ¼ë¡œ ë³€í™˜í•˜ì‹œì˜¤. datetimeê³¼ seasonì„ ì‹ë³„ì(id_vars)ë¡œ ìœ ì§€í•˜ê³ , casualê³¼ registeredë¥¼ ê°’ ì¸ì(value_vars)ë¡œ í•˜ì—¬ ë°ì´í„° í”„ë ˆì„ì„ ìƒì„±í•˜ì‹œì˜¤.
#melt

df_bike_long = df_bike.melt(id_vars=['datetime', 'season'], value_vars=['casual', 'registered'], var_name='íšŒì›ì—¬ë¶€', value_name = 'ëŒ€ì—¬ìˆ˜')
print(df_bike_long.head())
```
